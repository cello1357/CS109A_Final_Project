# DEMOGRAPHICS BY STATE AND DISTRICT

```python
import os
import glob

import numpy as np
import pandas as pd
```

```python
demograhics_df = pd.read_csv("data/demographics_by_district.csv", index_col=0, encoding='utf-8-sig')
demograhics_labels = demograhics_df.iloc[[0,1,2],:] #get labels index

demographics_year = demograhics_df.iloc[3:,0] # get year 
demographics_year = demographics_year.astype('int') #specify data types of data set 

demographics_state_labels = demograhics_df.iloc[3:,1] # get state labels
demographics_state_labels = demographics_state_labels.astype('str') #specify data types of data set 

demographics_district_num = demograhics_df.iloc[3:,2] # get district number
demographics_district_num = demographics_district_num.astype('int') #specify data types of data set 

demograhics_subframe = demograhics_df.iloc[3:,3:] #get data set 
demograhics_subframe = demograhics_subframe.astype('float') #specify data types of data set 
    
demograhics_df = pd.concat([demographics_year, demographics_state_labels, demographics_district_num, demograhics_subframe], axis=1, ignore_index=True)
```

```python
display(demograhics_labels)
display(demograhics_df)
display(demograhics_df.describe())
```

```python
col_range = range(4,307,1)

demograhics_by_state = demograhics_df.groupby(
   [1]
).aggregate(
    {
        col_range: 'sum'  
    }
)
```
